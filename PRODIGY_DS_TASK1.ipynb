# Total population growth from 2001 to 2022 for each country
___________

```python
# Keep only needed columns and drop missing values
df_growth = df[['Country Name', 'Country Code', '2001', '2022']].dropna()

# Calculate absolute growth
df_growth['Population Growth (2001–2022)'] = df_growth['2022'] - df_growth['2001']

# Sort by highest growth
df_growth_sorted = df_growth.sort_values(by='Population Growth (2001–2022)', ascending=False)

# Show top 10 countries by growth
top_growth = df_growth_sorted.head(10)

print("Top 10 Countries by Population Growth (2001 to 2022):\n")
top_growth[['Country Name', 'Country Code', '2001', '2022', 'Population Growth (2001–2022)']]
```

```python
# plot
plt.figure(figsize=(10, 6))
plt.barh(top_growth['Country Name'], top_growth['Population Growth (2001–2022)'], color='seagreen')
plt.xlabel('Population Growth (in millions)')
plt.title('Top 10 Countries by Population Growth (2001–2022)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

____________
# Histogram
____________

```python
total_population = df[df['Series Code'] == 'SP.POP.TOTL'].dropna(subset=['2022'])

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(total_population['2022'], bins=15, color='salmon', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Histogram of Country Populations (2022)')
plt.tight_layout()
plt.show()
```

# Top 10 Countries by Population (2022)

```python
# Filter for total population series
total_pop_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Remove rows with missing 2022 values
total_pop_data_clean = total_pop_data.dropna(subset=['2022'])

# Sort by 2022 population and get top 10 countries
top_10_countries = total_pop_data_clean.sort_values(by='2022', ascending=False).head(10)

# Display result
print("Top 10 Countries by Total Population (2022):\n")
top_10_countries[['Country Name', 'Country Code', '2022']]
```

___________
# Stacked Bar Chart
________

_______
# Top 10 Countries by Male & Female Population (2022)
_________

```python
# Merge male and female population data
male_clean = df_male[['Country Code', 'Country Name', '2022']].rename(columns={'2022': 'Male Population'})
female_clean = df_female[['Country Code', '2022']].rename(columns={'2022': 'Female Population'})
```

```python
combined_pop = pd.merge(male_clean, female_clean, on='Country Code')
```

```python
# Add total population and sort
combined_pop['Total Population'] = combined_pop['Male Population'] + combined_pop['Female Population']
combined_top10 = combined_pop.sort_values(by='Total Population', ascending=False).head(10)
```

```python
print("\nTop 10 Countries by Combined Male & Female Population (2022):\n")
combined_top10[['Country Name', 'Country Code', 'Male Population', 'Female Population', 'Total Population']]
```

```python
# Select top 10 countries by total population
countries = combined_top10['Country Name']
male_pop = combined_top10['Male Population']
female_pop = combined_top10['Female Population']

# Set bar width and positions
bar_width = 0.4
index = np.arange(len(countries))

# Plot
plt.figure(figsize=(12, 6))
plt.bar(index, male_pop, bar_width, label='Male Population', color='steelblue')
plt.bar(index + bar_width, female_pop, bar_width, label='Female Population', color='salmon')

# Labels and legend
plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.title('Top 10 Countries by Male and Female Population (2022) - Stacked Bar Chart')
plt.xticks(index + bar_width / 2, countries, rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```

__________
# Top 10 Countries by Male and Female Population (2022)
__________

```python
# Sort and select top 10
combined_top10 = combined_pop.sort_values(by='Total Population', ascending=False).head(10)

countries_top = combined_top10['Country Name']
male_top = combined_top10['Male Population']
female_top = combined_top10['Female Population']
x_top = np.arange(len(countries_top))


# Plot
plt.figure(figsize=(12, 6))
plt.bar(x_top, male_top, label='Male Population', color='red')
plt.bar(x_top, female_top, bottom=male_top, label='Female Population', color='blue')

plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.title('Top 10 Countries by Male and Female Population (2022) - Stacked Bar Chart')
plt.xticks(x_top, countries_top, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Sort and select bottom 10
combined_bottom10 = combined_pop.sort_values(by='Total Population', ascending=True).head(10)

countries_bottom = combined_bottom10['Country Name']
male_bottom = combined_bottom10['Male Population']
female_bottom = combined_bottom10['Female Population']
x_bottom = np.arange(len(countries_bottom))

# Plot
plt.figure(figsize=(12, 6))
plt.bar(x_bottom, male_bottom, label='Male Population', color='red')
plt.bar(x_bottom, female_bottom, bottom=male_bottom, label='Female Population', color='blue')

plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.title('Lowest 10 Countries by Male and Female Population (2022) - Stacked Bar Chart')
plt.xticks(x_bottom, countries_bottom, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
